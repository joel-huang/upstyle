```python
import os
import shutil
from os import path
from random import randint
import math
```

```python
IMGS_PATH = r'D:\Upstyle\upstyle\img'
TEST_PATH = r'D:\Upstyle\upstyle\test'
```

```python
os.chdir(IMGS_PATH)
classes_list = [name for name in os.listdir(".") if os.path.isdir(name)]
print(classes_list)
```

These are all the classes we have. Let's take a look at the number of images we have per class.

```python
num_classes = {}
for i in range(0, len(classes_list)):
    full_path = IMGS_PATH + "/{}".format(classes_list[i])
    print(full_path, ":", len([name for name in os.listdir(full_path)]))
    num_classes[classes_list[i]] = len([name for name in os.listdir(full_path)])
```

```python
print(len(num_classes.values()))
```

```python
min_imgs = min(num_classes.values())
max_imgs = max(num_classes.values())
min_img_classes = []
for k, v in num_classes.items():
    if v == min_imgs:
        min_img_classes.append(k)
        print(k, v)
    elif v == max_imgs:
        print(k,v)
    else:
        continue
```

```python
print("Average Number of imgs per class = {}".format(sum(num_classes.values())/len(num_classes.values())))
print("Mode of Number of imgs per class = {}".format(max(set(num_classes.values()),key=list(num_classes.values()).count)))
```

So we have 5620 classes in total, and 51.46 images per class on average, and the most common number of images is 37. That's very small. Also, there are some classes with only one image such as the cropped faux fur coat, while you have other classes with 179 images, so there's some data imbalance as well. What I'll do is remove those classes with only 1 image in them, and move them to a folder called redundant.

```python
print(min_img_classes)
REDUNDANT_PATH = r'D:\Upstyle\upstyle\redundant'
os.chdir(REDUNDANT_PATH)
for i in range(0,len(min_img_classes)):
    path = IMGS_PATH + "\{}".format(min_img_classes[i])
    dest_path = REDUNDANT_PATH + "\{}".format(min_img_classes[i])
    os.mkdir(dest_path)
    shutil.move(path, dest_path)
```

```python
os.chdir(IMGS_PATH)
classes_list = [name for name in os.listdir(".") if os.path.isdir(name)]
num_classes = {}
for i in range(0, len(classes_list)):
    full_path = IMGS_PATH + "/{}".format(classes_list[i])
    num_classes[classes_list[i]] = len([name for name in os.listdir(full_path)])
    
print(len(num_classes.keys()))
```

So now there are 4 less images because we moved the 4 one-single image folders into the redundant folder. Let's check the minimum number of images now.

```python
min_imgs = min(num_classes.values())
max_imgs = max(num_classes.values())
min_img_classes = []
for k, v in num_classes.items():
    if v == min_imgs:
        min_img_classes.append(k)
        print(k, v)
    elif v == max_imgs:
        print(k,v)
    else:
        continue
```

So, there are still folders with like only 2 images in them. I don't think that's gonna work. What I'll do is define a min threshold of the number of images we need per class, and move all folders that don't meet this threshold to go into the redundant folder.

```python
MIN_THRESHOLD = 10
small_image_folders = {}
for k, v in num_classes.items():
    if v < MIN_THRESHOLD:
        small_image_folders[k] = v

print(small_image_folders)
print("Number of remaining folders would be = {}".format(len(num_classes.keys()) - len(small_image_folders.keys())))
```

```python
os.chdir(REDUNDANT_PATH)
for i in range(0,len(small_image_folders.keys())):
    path = IMGS_PATH + "\{}".format(list(small_image_folders.keys())[i])
    dest_path = REDUNDANT_PATH + "\{}".format(list(small_image_folders.keys())[i])
    os.mkdir(dest_path)
    shutil.move(path, dest_path)
    
print([name for name in os.listdir(".") if os.path.isdir(name)])
```

```python
os.chdir(IMGS_PATH)
classes_list = [name for name in os.listdir(".") if os.path.isdir(name)]
num_classes = {}
for i in range(0, len(classes_list)):
    full_path = IMGS_PATH + "/{}".format(classes_list[i])
    num_classes[classes_list[i]] = len([name for name in os.listdir(full_path)])
    
print(len(num_classes.keys()))
```

So now the remaining number of classes in the imgs folder is 5553, as expected, since we removed those folders that didn't meet the min threshold. Now, let's create the test set.

```python
os.chdir(TEST_PATH)
for name in classes_list:
    filename = TEST_PATH + "/{}".format(name)
    os.mkdir(filename)
    
print([name for name in os.listdir(".") if os.path.isdir(name)])
```

Now let's move some of the images over from the imgs folder to the test folder.

```python
TRAIN_TEST_SPLIT = 0.10
move_imgs = {}
for i in range(0, len(classes_list)):
    move_imgs[classes_list[i]] = []
    path = IMGS_PATH + "/{}".format(classes_list[i])
    filenames = os.listdir(path)
    for j in range(math.ceil(TRAIN_TEST_SPLIT * len(filenames))):
        randidx = randint(0,len(filenames))
        move_imgs[classes_list[i]].append(randidx)
    
print(move_imgs)
```

So to ensure that we have a representation of all classes, we will take 10% of the number of images in each class and move them into the respective folder in test. move_imgs is a dict containing _classname_: _list of indexes to be moved_ 

```python
for i in range(0, len(classes_list)):    
    src_path = IMGS_PATH + "/{}".format(classes_list[i])
    dest_path = TEST_PATH + "/{}".format(classes_list[i])
    os.chdir(src_path)
    filenames = os.listdir(src_path)
    move_list = move_imgs[classes_list[i]]
    for j in range(0, len(move_list)-1):
        idx = move_list[j]
        shutil.copy(filenames[idx], dest_path)
```
