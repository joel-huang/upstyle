```python
import pandas as pd
import numpy as np

directory = '/home/joel/Downloads/'
df_path = directory + 'anno/attr_zeroed_upper_lower.csv'

frames = [chunk for chunk in pd.read_csv(df_path, chunksize=20000, low_memory=False, index_col=None)]
df = pd.concat(frames)
```

```python
df.tail()
```

```python
df_upper = df[df['upper_lower']==1]
df_lower = df[df['upper_lower']==2]
```

```python
# Get 1000 labels and their type (ranges from 1-5)
def get_attr_type():
    attr_type = {}
    attr_type_file = open(directory + 'anno/list_attr_cloth.txt', 'r')
    lines = attr_type_file.readlines()
    for i in lines[2:]:
        s = i.split()
        attr = s[:-1]
        attr = ' '.join(attr)
        attr_num = s[-1]
        attr_type[attr] = attr_num
    return attr_type

attr_type = get_attr_type()
df_attr = pd.DataFrame()
df_attr = df_attr.from_dict(attr_type.items())
df_attr.columns = ['Attribute','Type']
```

```python
d_texture = []
d_fabric = []
d_shape = []
d_part = []
d_style = []

for i,r in df_attr.iterrows():
    x = r["Type"]
    y = r['Attribute']
    if x == "1":
        d_texture.append(y)
    elif x == "2":
        d_fabric.append(y)
    elif x == "3":
        d_shape.append(y)
    elif x == "4":
        d_part.append(y)
    else:
        d_style.append(y)
        
d_texture.insert(0, 'image_name')
d_fabric.insert(0, 'image_name')
d_shape.insert(0, 'image_name')
d_part.insert(0, 'image_name')
d_style.insert(0, 'image_name')
```

```python
# Column reduction from 1000 to the various 5 types of attributes.

df_texture_u = df_upper[d_texture]
df_fabric_u = df_upper[d_fabric]
df_shape_u = df_upper[d_shape]
df_part_u = df_upper[d_part]
df_style_u = df_upper[d_style]

df_texture_l = df_lower[d_texture]
df_fabric_l = df_lower[d_fabric]
df_shape_l = df_lower[d_shape]
df_part_l = df_lower[d_part]
df_style_l = df_lower[d_style]
```

```python
# Produce uncollapsed CSVs with file names
df_texture_u.to_csv(directory + 'anno/processed_texture_upper.csv', index=False)
df_fabric_u.to_csv(directory + 'anno/processed_fabric_upper.csv', index=False)
df_shape_u.to_csv(directory + 'anno/processed_shape_upper.csv', index=False)
df_part_u.to_csv(directory + 'anno/processed_part_upper.csv', index=False)
df_style_u.to_csv(directory + 'anno/processed_style_upper.csv', index=False)

df_texture_l.to_csv(directory + 'anno/processed_texture_lower.csv', index=False)
df_fabric_l.to_csv(directory + 'anno/processed_fabric_lower.csv', index=False)
df_shape_l.to_csv(directory + 'anno/processed_shape_lower.csv', index=False)
df_part_l.to_csv(directory + 'anno/processed_part_lower.csv', index=False)
df_style_l.to_csv(directory + 'anno/processed_style_lower.csv', index=False)
```
